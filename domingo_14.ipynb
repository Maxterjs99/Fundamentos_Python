### Edades faltantes - clase 1

```python
df_age_c1 = df2[ (df2['Pclass'] == 1) & (df2['Age'].isnull()) ]
```

```python
df_age_c1_female = df_age_c1[df_age_c1['Sex'] == 'female']
df_age_c1_male = df_age_c1[df_age_c1['Sex'] == 'male']
```

```python
df_age_c1_male.head()
```

### Edades faltantes - clase 2

```python
df_age_c2 = df2[ (df2['Pclass'] == 2) & (df2['Age'].isnull())]
```

```python
df_age_c2_female = df_age_c2[df_age_c2['Sex'] == 'female']
df_age_c2_male = df_age_c2[df_age_c2['Sex'] == 'male']
```

### Edades faltantes - clase 3

```python
df_age_c3 = df2[ (df2['Pclass'] == 3) & (df2['Age'].isnull())]
```

```python
df_age_c3_female = df_age_c3[df_age_c3['Sex'] == 'female']
df_age_c3_male = df_age_c3[df_age_c3['Sex'] == 'male']
```

```python
print(f'Edades faltantes clase 1, sexo femenino: {df_age_c1_female.shape[0]}')
print(f'Edades faltantes clase 1, sexo masculino: {df_age_c1_male.shape[0]}')
print(f'Edades faltantes clase 2, sexo femenino: {df_age_c2_female.shape[0]}')
print(f'Edades faltantes clase 2, sexo masculino: {df_age_c2_male.shape[0]}')
print(f'Edades faltantes clase 3, sexo femenino: {df_age_c3_female.shape[0]}')
print(f'Edades faltantes clase 3, sexo masculino: {df_age_c3_male.shape[0]}')
```

```python
df_prom_age
```

```python
prom_female_1 = df_prom_age.loc[(1, 'female'), 'mean']
prom_male_1 = df_prom_age.loc[(1, 'male'), 'mean']
prom_female_2 = df_prom_age.loc[(2, 'female'), 'mean']
prom_male_2 = df_prom_age.loc[(2, 'male'), 'mean']
prom_female_3 = df_prom_age.loc[(3, 'female'), 'mean']
prom_male_3 = df_prom_age.loc[(3, 'male'), 'mean']
```

```python
df_age_c1_female['Age'] = df_age_c1_female['Age'].fillna(round(prom_female_1, 1))
df_age_c1_male['Age'] = df_age_c1_male['Age'].fillna(round(prom_male_1, 1))
df_age_c2_female['Age'] = df_age_c2_female['Age'].fillna(round(prom_female_2, 1))
df_age_c2_male['Age'] = df_age_c2_male['Age'].fillna(round(prom_male_2, 1))
df_age_c3_female['Age'] = df_age_c3_female['Age'].fillna(round(prom_female_3, 1))
df_age_c3_male['Age'] = df_age_c3_male['Age'].fillna(round(prom_male_3, 1))
```

```python
df_age_c2_female
```

```python
# delete NaN in age inside dataframe df2
df3 = df2.dropna(subset=['Age'])
df3.info()
```

```python
# Concatenar varios dataframe en uno solo

df_prom_age10 = pd.concat([
    df_age_c1_female, 
    df_age_c1_male,
    df_age_c2_female, 
    df_age_c2_male,
    df_age_c3_female, 
    df_age_c3_male
])

df_prom_age10
```

```python
df4 = pd.concat([
    df3,
    df_prom_age10
    ])

df4 = df4.reset_index(drop=True)
df4
```
